## 입력변수 2개 이상

* 예습시간 x1, 복습시간 x2와 pass fail 상관관계

```python
import numpy as np

x_data = np.array([[2,4],[4,11],[6,6],[8,5],[10,7],[12,16],[14,8],[16,3],[18,7]])
t_data = np.array([0,0,0,0,1,1,1,1,1]).reshape(9,1)

W = np.random.rand(2,1) # 2x1 행렬
b = np.random.rand(1)

print(W, W.shape, b, b.shape)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def loss_func(x, t):
    
    delta = 1e-7 # log 무한대 발산 방지(입력값 0 안되게)
    
    z = np.dot(x,W) + b
    y = sigmoid(z)
    
    # cross-entropy
    return -np.sum(t*np.log(y + delta) + (1-t)*np.log((1-y)+delta))
```

```python
def numerical_derivative(f,x):
    delta_x = 1e-4 # 0.0001
    grad = np.zeros_like(x)
    
    it = np.nditer(x, flags = ['multi_index'], op_flags = ['readwrite'])
    
    while not it.finished:
        idx = it.multi_index
        tmp_val = x[idx]
        x[idx] = float(tmp_val) + delta_x
        fx1 = f(x) # f(x+delta_x)
        
        x[idx]  = tmp_val - delta_x
        fx2 = f(x) # f(x-delta_x)
        grad[idx] = (fx1 - fx2) / (2*delta_x)
        
        x[idx] = tmp_val
        it.iternext()
        
    return grad
        
        
        
```

```python
def predict(x):
    
    z = np.dot(x,W) + b
    y = sigmoid(z)
    
    if y > 0.5:
        result = 1 #True
    else:
        result = 0 #False
        
    return y, result
```

```python
learning_rate = 1e-2 # 발산하는 경우, 1e-3, 1e-6 등으로 바꾸어서 실행

f = lambda x : loss_func(x_data, t_data) # f(x) = loss_func(x_data, t_data)

print("initial error value = ", loss_func(x_data,t_data), "initial W = ", W, "\n", " , b = ", b)

for step in range(80001):
    
    W -= learning_rate*numerical_derivative(f, W)
    b -= learning_rate*numerical_derivative(f, b)
    
    if(step%400 == 0):
        print("step = ", step, "error value = ", loss_func(x_data, t_data), "W = ", W, ", b = ", b)
```

```python
test_data = np.array([3,17]) # (예습, 복습) = (3,17)
predict(test_data)
```
